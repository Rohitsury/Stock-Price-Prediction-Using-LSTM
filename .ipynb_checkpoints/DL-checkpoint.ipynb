```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from pandas_datareader import data as pdr
from datetime import datetime
```

```python
yf.pdr_override()
start = '2010-1-1'
end = '2023-4-5'
df = pdr.get_data_yahoo('TCS.NS', start=start,end=end)
df = df.reset_index()
```

```python
plt.figure(figsize=(16,8))
plt.title('Apple')
plt.xlabel('Days')
plt.ylabel('Opening Price USD ($)')
plt.plot(df['Open'])
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title('Apple')
plt.xlabel('Days')
plt.ylabel('High Price USD ($)')
plt.plot(df['High'])
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title('Apple')
plt.xlabel('Days')
plt.ylabel('Low Price USD ($)')
plt.plot(df['Low'])
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title('Apple')
plt.xlabel('Days')
plt.ylabel('Closing Price USD ($)')
plt.plot(df['Close'])
plt.show()
```

```python
df1 = df['Open']
df2 = df['Close']
df3 = df['High']
df4 = df['Low']
```

```python
df1 = pd.DataFrame(df1) 
df2 = pd.DataFrame(df2)   

df3 = pd.DataFrame(df3) 
df4 = pd.DataFrame(df4)
```

```python


# Prediction 100 days into the future.
future_days = 1
df1['Prediction_Open'] = df1['Open'].shift(-future_days)
df2['Prediction_Close'] = df2['Close'].shift(-future_days)
df3['Prediction_High'] = df3['High'].shift(-future_days)
df4['Prediction_Low'] = df4['Low'].shift(-future_days)


```

```python

x_open = np.array(df1.drop(['Prediction_Open'], 1))[:-future_days]
x_close = np.array(df2.drop(['Prediction_Close'], 1))[:-future_days]

x_high = np.array(df3.drop(['Prediction_High'], 1))[:-future_days]
x_low = np.array(df4.drop(['Prediction_Low'], 1))[:-future_days]
print(x_open)
print(x_close)
print(x_high)
print(x_low)


```

```python
y_open = np.array(df1['Prediction_Open'])[:-future_days]
y_close = np.array(df2['Prediction_Close'])[:-future_days]
y_high = np.array(df3['Prediction_High'])[:-future_days]
y_low = np.array(df4['Prediction_Low'])[:-future_days]
print(y_open)
print(y_close)
print(y_high)
print(y_low)
```

```python


from sklearn.model_selection import train_test_split
x_open_train, x_open_test, y_open_train, y_open_test = train_test_split(x_open, y_open, test_size = 0.2)
x_close_train, x_close_test, y_close_train, y_close_test = train_test_split(x_close, y_close, test_size = 0.2)

x_high_train, x_high_test, y_high_train, y_high_test = train_test_split(x_high, y_high, test_size = 0.2)

x_low_train, x_low_test, y_low_train, y_low_test = train_test_split(x_low, y_low, test_size = 0.2)



```

```python


from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression

```

```python
# Implementing Linear and Decision Tree Regression Algorithms.
tree_open = DecisionTreeRegressor().fit(x_open_train, y_open_train)
tree_close = DecisionTreeRegressor().fit(x_close_train, y_close_train)
tree_high = DecisionTreeRegressor().fit(x_high_train, y_high_train)
tree_low = DecisionTreeRegressor().fit(x_low_train, y_low_train)

lr_open = LinearRegression().fit(x_open_train, y_open_train)
lr_close = LinearRegression().fit(x_close_train, y_close_train)
lr_high = LinearRegression().fit(x_high_train, y_high_train)
lr_low = LinearRegression().fit(x_low_train, y_low_train)
```

```python
x_future_open = df1.drop(['Prediction_Open'], 1)[:-future_days]
x_future_open = x_future_open.tail(future_days)
x_future_open = np.array(x_future_open)
print(x_future_open)

x_future_close = df2.drop(['Prediction_Close'], 1)[:-future_days]
x_future_close = x_future_close.tail(future_days)
x_future_close = np.array(x_future_close)
print(x_future_close)

x_future_high = df3.drop(['Prediction_High'], 1)[:-future_days]
x_future_high = x_future_high.tail(future_days)
x_future_high = np.array(x_future_high)
print(x_future_high)

x_future_low = df4.drop(['Prediction_Low'], 1)[:-future_days]
x_future_low = x_future_low.tail(future_days)
x_future_low = np.array(x_future_low)
x_future_low
```

```python
tree_prediction_open = tree_open.predict(x_future_open)
print(tree_prediction_open)

tree_prediction_close = tree_close.predict(x_future_close)
print(tree_prediction_close)

tree_prediction_high = tree_high.predict(x_future_high)
print(tree_prediction_high)

tree_prediction_low = tree_low.predict(x_future_low)
print(tree_prediction_low)
```

```python
lr_prediction_open = lr_open.predict(x_future_open)
print(lr_prediction_open)
lr_prediction_close = lr_close.predict(x_future_close)
print(lr_prediction_close)
lr_prediction_high = lr_high.predict(x_future_high)
print(lr_prediction_high)
lr_prediction_low = lr_low.predict(x_future_low)
print(lr_prediction_low)
```

```python


predictions_open = tree_prediction_open 
valid1 = df1[x_open.shape[0]:]
valid1['Predictions_Open'] = predictions_open


predictions_close = tree_prediction_close 
valid2 = df2[x_close.shape[0]:]
valid2['Predictions_Close'] = predictions_close

predictions_high = tree_prediction_high 
valid3 = df3[x_high.shape[0]:]
valid3['Predictions_High'] = predictions_high



predictions_low = tree_prediction_low
valid4 = df4[x_low.shape[0]:]
valid4['Predictions_low'] = predictions_low
 
```

```python
plt.figure(figsize=(16,8))
plt.title("Model")
plt.xlabel('Days')
plt.ylabel('Close Price USD ($)')
plt.plot(df2['Close'])
plt.plot(valid2[['Close', 'Predictions_Close']])
plt.legend(["Original", "Valid", 'Predicted'])
plt.show()
```
